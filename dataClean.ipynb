```python
import pandas as pd

mvps = pd.read_csv('mvps.csv')

mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]

mvps
```

```python
players = pd.read_csv("players.csv")

del players["Unnamed: 0"]
del players["Rk"]

players["Player"] = players["Player"].str.replace("*", "", regex=False)

players.head(20)
```

```python
def single_row(df):
    if df.shape[0]== 1: #If player played on one team return the row
        return df
    else:
        row = df[df["Tm"] == "TOT"] #If multiple teams, take TOT stats
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players.head(20)
```

```python
combined = players.merge(mvps, how='outer', on=['Player', 'Year']) #Combine Players and MVP tables into one using Player and Year as index
```

```python
combined[combined['Pts Won'] > 0]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
combined
```

```python
teams = pd.read_csv('teams.csv')
teams.head(30)
```

```python
teams = teams[~teams['W'].str.contains('Division')]
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex = False)
```

```python
teams.head()
```

```python
teams["Team"].unique()
```

```python
combined['Tm'].unique()
```

```python
import csv
nicknames = {}

with open ('nicknames.txt') as file:
    reader = csv.DictReader(file)
    for team in reader:
        nicknames.update({team['Abbreviation']: team["Name"]})     
print(nicknames)
```

```python
combined["Team"] = combined['Tm'].map(nicknames)
combined.head(5)
```

```python
stats = combined.merge(teams, how='outer', on=['Team', 'Year'])
stats
```

```python
del stats['Unnamed: 0']
stats
```

```python
stats = stats.apply(pd.to_numeric, errors='ignore')
stats.dtypes
```
